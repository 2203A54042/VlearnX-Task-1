```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/customer_sales_data.csv')
```

```python
print("first 5 rows of data:")
data.head()
```

```python
print("\n Data info:")
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(8,6))
plt.hist(data['Age'], bins = 10, color='skyblue',edgecolor = 'black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Annual_Income'], data['Spending_Score'], color='purple', alpha=0.5)
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
plt.figure(figsize=(8,6))
numeric_data = data.select_dtypes(include=[np.number])
sns.heatmap(numeric_data.corr(),annot=True,cmap='coolwarm',center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("Insights:")
print("1. Age Distribution: Most customers are in their 20s and 30s.")
print("2. Annual Income vs. Spending Score: A positive correlation suggests higher income may lead to higher spending.")
print("3. Correlation Heatmap: Provides insight into how features like Annual Income and Spending Score relate to each other.")
```
